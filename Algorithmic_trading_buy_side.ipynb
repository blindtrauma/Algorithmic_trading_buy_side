```python

import numpy as np
#An alternate strategy to calculate adx can be TA-Lib library which can be used and levraged for technical analysis in algorithmic trading

def calculate_adx(high, low, close, period):

  if not all(isinstance(arr, np.ndarray) for arr in [high, low, close]):
    raise ValueError("All input features (high, low, close) must be NumPy arrays.")

  high = high[-period:]

  low = low[-period:]
  close = close[-period:]



  high_low = high - low
  high_close_prev = np.abs(high - np.roll(close, 1))
  low_close_prev = np.abs(low - np.roll(close, 1))
  tr = np.maximum(high_low, high_close_prev)
  tr = np.maximum(tr, low_close_prev)


  move_up = high - np.roll(high, 1)
  move_down = np.roll(low, 1) - low

  pos_dm = np.where((move_up > move_down) & (move_up > 0), move_up, 0)
  neg_dm = np.where((move_down > move_up) & (move_down > 0), move_down, 0)

  pos_di = 100 * np.cumsum(pos_dm) / np.cumsum(tr)
  neg_di = 100 * np.cumsum(neg_dm) / np.cumsum(tr)

  # Step 3: Calculate the Directional Index (DI) and Average Directional Index (ADX)
  dx = 100 * np.abs((pos_di - neg_di) / (pos_di + neg_di))
  print(dx)
  adx = np.mean(dx)

  return adx



high = data30m['High']
high = np.array(high)
low = data30m['Low']
low = np.array(low)
close = data30m['Close']
close = np.array(close)
adxisequalto= calculate_adx(high, low, close, 182)
print(adxisequalto)
```
